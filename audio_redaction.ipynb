```python
import whisper

model = whisper.load_model("small.en")
result = model.transcribe("Recording (3).m4a", word_timestamps=True)
```

```python
import spacy
# Load SpaCy for NER
nlp = spacy.load("en_core_web_sm")

# Process transcription and extract entities with timestamps
entities_to_redact = {"PERSON", "EMAIL", "ORG","LOCATION"}  # User-defined categories to redact
sensitive_segments = []

for chunk in result["chunks"]:
    text = chunk["text"]
    doc = nlp(text)
    for ent in doc.ents:
        if ent.label_ in entities_to_redact:
            sensitive_segments.append({
                "entity": ent.text,
                "start": chunk["timestamp"][0],
                "end": chunk["timestamp"][1]
            })
```
